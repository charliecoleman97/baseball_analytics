```python
import pybaseball as pb
import pandas as pd
import plotly.express as px
import plotly.io as pio
import numpy as np
from PIL import Image
```

## Plotting Contact pct vs AVG

```python
''' Going to use 2021 batting stats with a minimum of 200 PA'''
pd.set_option('display.max_columns', None)
df = pb.batting_stats(2021, qual=200)
df
```

### High Level Anaylsis

```python
''' Selecting the columns we need'''
plot_df = (
    df
     .pipe(lambda df: df[["IDfg", "Name", "Team", 'G', 'PA','AVG', 'SLG', 'OPS', 'Dol', 'Spd','Off', 'WAR', 'wRC+', 'BsR', 'BB%','HardHit%', 'BABIP', 'Barrel%', 'Contact%', 'GB%', 'FB%']]) # just getting the columns we need
)
```

```python
plot_df['HardHit%'] = plot_df['HardHit%'] * 100
plot_df['BB%'] = plot_df['BB%'] * 100
plot_df['Barrel%'] = plot_df['Barrel%'] * 100
plot_df['Contact%'] = plot_df['Contact%'] * 100
plot_df['GB%'] = plot_df['GB%'] * 100
plot_df['FB%'] = plot_df['FB%'] * 100
```

```python
plot_df['Contact%'].describe()
```

```python
plot_df['AVG'].describe()
```

```python
'''Plotting Scatter'''
plt = px.scatter(
    plot_df, x="Contact%", y="AVG", 
    trendline='ols',
    trendline_color_override="purple",
    color="wRC+",
    hover_name="Name",
    hover_data=["PA", "BABIP", "HardHit%", 'GB%', 'FB%'],
    color_continuous_scale=px.colors.diverging.Tropic,
    width=1500, height=800,
)

plt.update_traces(marker_line_width=2, marker_size=10)
plt.update_yaxes(nticks=20)
plt.update_xaxes(nticks=20)


```
