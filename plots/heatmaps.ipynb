```python
from pybaseball import schedule_and_record
import pandas as pd
```

```python
df = pd.read_csv('sportsref_download.csv')
df
```

```python
col_to_keep = ['SP', 'RP', 'C', '1B', '2B', '3B', 'SS', 'LF', 'CF', 'RF', 'DH', 'PH']

trimmed_df = df[col_to_keep]
trimmed_df
```

```python
def split_col(col):
    return [col.str.slice(0,3), col.str.slice(3)]
```

```python
def melt_war_table(df, cols):
    dfs = []
    for pos in cols:
        col_team, col_val = split_col(df[pos])
        temp_df = pd.DataFrame({'team': col_team, 'war': col_val})
        temp_df['position'] = pos
        temp_df['war'] = pd.to_numeric(temp_df['war'])
        dfs.append(temp_df)
    return pd.concat(dfs, ignore_index=True)

```

```python
melted_df = melt_war_table(df, col_to_keep)
melted_df
```
