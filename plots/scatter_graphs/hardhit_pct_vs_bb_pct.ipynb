```python
import pybaseball as pb
import pandas as pd
import plotly.express as px
import plotly.io as pio
import numpy as np
from PIL import Image
```

### Plotting Hard Hit % vs BB%

```python
pd.set_option('display.max_columns', None)
```

```python
''' Going to use 2021 batting stats with a minimum of 200 PA'''
df = pb.batting_stats(2021, qual=200)
df
```

```python
''' Selecting the columns we need'''
plot_df = (
    df
     .pipe(lambda df: df[["IDfg", "Name", "Team", 'G', 'PA', 'OPS', 'Dol', 'Spd','Off', 'WAR', 'wRC+', 'Def', 'BsR', 'BB%','HardHit%', 'BABIP', 'Barrel%']]) # just getting the columns we need
)
```

```python
plot_df['HardHit%'] = plot_df['HardHit%'] * 100
plot_df['BB%'] = plot_df['BB%'] * 100
plot_df['Barrel%'] = plot_df['Barrel%'] * 100
```

```python
plot_df
```

### High level analysis

```python
plot_df['HardHit%'].describe()
```

```python
plot_df['BB%'].describe()
```

```python
'''Plotting Scatter'''
plt = px.scatter(
    plot_df, x="Barrel%", y="BB%",
    color="wRC+",
    hover_name="Name",
    hover_data=["PA", "BABIP"],
    color_continuous_scale=px.colors.diverging.Tropic,
    width=1500, height=800,
)

plt.update_traces(marker_line_width=2, marker_size=10)
plt.update_yaxes(nticks=20)
plt.update_xaxes(nticks=40)


'''Medians'''
plt.add_hline(y=np.median(plot_df["BB%"]), line_dash="dot", row="all", col="all",
              annotation_text="Median BB%", 
              annotation_position="bottom right")

plt.add_vline(x=np.median(plot_df["Barrel%"]), line_dash="dot", row="all", col="all",
              annotation_text="Median Barrel%", 
              annotation_position="bottom right")

'''Plotting top 10th percentile '''
plt.add_shape(
    x0=plot_df["Barrel%"].quantile(q=0.9), 
    y0=plot_df["BB%"].quantile(q=0.9),
    x1=plot_df["Barrel%"].quantile(q=1),
    y1=plot_df["BB%"].quantile(q=1),
    line=dict(color="LightSeaGreen", width=2, dash="dot")
)
plt.add_annotation(
    x=plot_df["Barrel%"].quantile(q=0.99),
    y=plot_df["BB%"].quantile(q=0.999),
    text="Great Hitters",
    showarrow=False,
    font=dict(
        family="arial",
        size=18,
        color="black"
    )
    )


'''Plotting bottom 10th percentile'''
plt.add_shape(
    x0=plot_df["Barrel%"].quantile(q=0), 
    y0=plot_df["BB%"].quantile(q=0),
    x1=plot_df["Barrel%"].quantile(q=0.2),
    y1=plot_df["BB%"].quantile(q=0.2),
    line=dict(color="#EF553B", width=2, dash="dot")
)
plt.add_annotation(
    x=2, 
    y=2,
    text="Terrible Hitters",
    showarrow=False,
    font=dict(
        family="arial",
        size=18,
        color="black"
    )
    )

plt
```
