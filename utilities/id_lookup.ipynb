```python
import pybaseball as pb
from pybaseball import statcast_fielding
import pandas as pd
import plotly.express as px
import plotly.io as pio
import numpy as np
from PIL import Image
```

```python
df = statcast_fielding.statcast_outs_above_average(2021, "all", 1)
df.rename(columns={" first_name": "first_name"},inplace=True)
df["first_name"] = df["first_name"].str.lstrip()

name_df = (
    df
     .pipe(lambda df: df[['last_name','first_name']]) # just getting the columns we need
)
name_df

```

```python
name_list = name_df.values.tolist()
name_list
```

```python
id_df = pb.player_search_list(name_list)
# id_df.rename(columns={"key_fangraphs": "IDfg", "key_mlbam": "player_id"},inplace=True)
id_df
```

```python
hitting_df = pb.batting_stats(2021, qual=10)
hitting_df= (
    hitting_df
     .pipe(lambda df: df[["IDfg", "Name", "Team", 'G']]) # just getting the columns we need
)
```

```python
merged_df = pd.merge(hitting_df, id_df[['key_fangraphs', 'key_mlbam', 'key_bbref', 'key_retro']], left_on='IDfg', right_on= 'key_fangraphs', how='left')
```
